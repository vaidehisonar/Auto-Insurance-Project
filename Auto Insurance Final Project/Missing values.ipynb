```python
import pandas as pd # storing and mnipulating the dataset
import matplotlib.pyplot as plt # for data visualization 
import seaborn as sns # for data visualization 
import warnings  
warnings.filterwarnings("ignore") # to not display any warnings generated 
import numpy as np # for array computations
```

```python
df= pd.read_csv('training_data_BigData.csv') #df_assg1 holds the dataset
df.head() #displacing the first 5 records
```

```python
df = df[df.Vehicle_Days_Per_Week_Driven < 8] 
df=df[df.Vehicle_Days_Per_Week_Driven != -1 ]
df=df[df.Vehicle_Anti_Theft_Device != 'Unknown']
df=df[df.EEA_Policy_Zip_Code_3 != 'Unknown']
df=df[df.Vehicle_Passive_Restraint != 'Unknown' ]
#df=df[df.Vehicle_Bodily_Injury_Limit != 'NA']
df.Vehicle_Bodily_Injury_Limit.replace(to_replace ='1M-1M  ' , value = '1',inplace=True)
df=df[df.Vehicle_Bodily_Injury_Limit !='1']

df.Vehicle_Bodily_Injury_Limit.replace(to_replace ='500-500' , value = 'other',inplace=True)
df.Vehicle_Bodily_Injury_Limit.replace(to_replace ='300-500' , value = 'other',inplace=True)
df.Vehicle_Bodily_Injury_Limit.replace(to_replace ='100-500' , value = 'other',inplace=True)

df.EEA_Prior_Bodily_Injury_Limit.replace(to_replace ='300-300' , value = 'other',inplace=True)
df.EEA_Prior_Bodily_Injury_Limit.replace(to_replace ='200-400' , value = 'other',inplace=True)
df.EEA_Prior_Bodily_Injury_Limit.replace(to_replace ='75-300' , value = 'other',inplace=True)
df.EEA_Prior_Bodily_Injury_Limit.replace(to_replace ='750-750' , value = 'other',inplace=True)


#df.Vehicle_Bodily_Injury_Limit.replace(to_replace ='NA' , value = 'not-applicable',inplace=True)
df['EEA_Policy_Zip_Code_3'] = df['EEA_Policy_Zip_Code_3'].astype(int)
#df['Vehicle_Passive_Restraint'] = df['Vehicle_Passive_Restraint'].astype(int)

df=df.drop(['EEA_PolicyYear'],axis=1)
```

```python
from sklearn.linear_model import LinearRegression
x=df[(df['Vehicle_Symbol']!=-1)]
y_train=x[['Vehicle_Symbol']]
x_train=x[['Annual_Premium','Vehicle_Age_In_Years']]

#x_test=x_train[:20]
#x_train.drop(x_train.tail(20).index,inplace=True)

x1=df[(df['Vehicle_Symbol']==-1)]
x_test=x1[['Annual_Premium','Vehicle_Age_In_Years']]

y_test=x1[['Vehicle_Symbol']]
#y_train.drop(y_train.tail(20).index,inplace=True)

reg = LinearRegression().fit(x_train, y_train)
y_pred = reg.predict(x_test)

#print(y_test-y_pred)
y_pred=y_pred.astype(int)

new_df=df

new_df.loc[new_df.Vehicle_Symbol == -1, 'Vehicle_Symbol'] = y_pred
```
