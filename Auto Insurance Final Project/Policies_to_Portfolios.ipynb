```
from pandas import ExcelWriter
from pandas import ExcelFile
from scipy import stats
import os,glob
Driver_Total_Senior_Ages_65_69_mean=[]
Driver_Total_Teenager_Age_15_19_mean=[]
Driver_Total_Licensed_In_State_mode=[]
Driver_Total_mode=[]
Vehicle_Performance_High_count=[]
Vehicle_Youthful_Driver_Indicator_Y_count=[]
sys_renewed_n_0_count=[]
sys_renewed_n_1_count=[]
folder_path = '/content/gdrive/My Drive/bigdataproj/'
for filename in glob.glob(os.path.join(folder_path, '*.xlsx')):
    p = pd.read_excel(filename)
    #print(p.head())
    Driver_Total_Teenager_Age_15_19_mean.append(p['Driver_Total_Teenager_Age_15_19'].mean())
    Driver_Total_Senior_Ages_65_69_mean.append(p['Driver_Total_Senior_Ages_65_69'].mean())
    Driver_Total_Licensed_In_State_mode.append(stats.mode(p['Driver_Total_Licensed_In_State']))
    Driver_Total_mode.append(stats.mode(p['Driver_Total']))
    Vehicle_Youthful_Driver_Indicator_Y_count.append(p['Vehicle_Youthful_Driver_Indicator_Y'].sum())
    Vehicle_Performance_High_count.append(p['Vehicle_Performance_High'].sum())
    sys_renewed_n_0_count.append(p['sys_renewed_n_0'].sum())
    sys_renewed_n_1_count.append(p['sys_renewed_n_1'].sum())
    print(filename)
```
